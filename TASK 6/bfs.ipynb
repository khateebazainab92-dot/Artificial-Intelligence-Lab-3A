## BFS WITHOUT QUEUE

```python
def bfs_without_queue(graph, start):
    visited = []
    queue = []   # Using list as queue

    visited.append(start)
    queue.append(start)

    while queue:
        node = queue.pop(0)   # Remove first element
        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)


# Example Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

print("BFS Traversal (Without Queue):")
bfs_without_queue(graph, 'A')
```

## BFS WITH QUEUE

```python
from collections import deque

def bfs_with_queue(graph, start):
    visited = set()
    queue = deque([start])

    visited.add(start)

    while queue:
        node = queue.popleft()   # Remove from front
        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)


# Example Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

print("BFS Traversal (With Queue):")
bfs_with_queue(graph, 'A')
```
